```python
import os
import h5py
import json
import numpy as np
import pandas as pd
import nibabel as nib
from pathlib import Path

from scipy import stats
from sklearn import svm
from statsmodels.stats.multitest import multipletests
from scipy.stats import ttest_1samp
from sklearn.model_selection import cross_val_score, StratifiedKFold

%matplotlib inline
```

```python
from helper import decoding
from utils import filter_task_betas, beta_concatenation, make_brain_surface_plot, df_concatenation
```

```python
# load glasser atlas
parent_dir = Path.cwd().parent
glasser_atlas_str= os.path.join(parent_dir,'data', 'Glasser_LR_Dense64k.dlabel.nii')
glasser_atlas = nib.load(glasser_atlas_str).get_fdata()[0].astype(int)
print("Glassier Atlas shape:", glasser_atlas.shape)
num_regions = 360
```

```python
# Load data
subj = "sub-03"
excluded_session = 1000  # For sanity check purposes
sessions = [f"ses{i}" for i in range(1, 17) if i != excluded_session]

# Define run identifiers
runs = [f"run-{i:02d}" for i in range(1, 6)]

# Specify directory
datadir = (
    f"{parent_dir}/data/{subj}/glm_betas/{subj}/glm_betas_encoding_delay_full_TR_betas/{subj}/"
)  # Dataset 1: includes all delay periods
```

```python
# Initialize data structures
task_betas = {}
df_conditions = {}

# List of tasks
# tasks = [
#     'dmsloc', '1backloc', '1backctg', '1backobj', 'ctxlco', 'ctxcol', 
#     'interdmsobjABAB', 'interdmslocABBA', 'interdmslocABAB',
#     'interdmsctgABAB', 'interdmsobjABBA', 'interdmsctgABBA'
# ]
tasks = ['interdmsctgABAB', 'interdmsctgABBA']

# Iterate over all task-session-run combinations
for task in tasks:
    print(f"Processing task: {task}")
    task_betas[task] = {}
    df_conditions[task] = {}

    for sess in sessions:
        task_betas[task][sess] = {}
        df_conditions[task][sess] = {}

        for run in runs:
            # Construct file paths
            base_filename = f"glmmethod1_{sess}_task-{task}_{run}"
            h5_file_path = os.path.join(datadir, f"{base_filename}_betas.h5")
            csv_file_path = os.path.join(datadir, f"{base_filename}.csv")

            # Check if HDF5 file exists before processing
            if not os.path.exists(h5_file_path):
                print(f"Warning: HDF5 file not found: {h5_file_path}")
                continue

            print(f"------------Loading: {h5_file_path}-------------")
            try:
                with h5py.File(h5_file_path, 'r') as h5f:
                    task_betas[task][sess][run] = h5f['betas'][:].copy()

                if os.path.exists(csv_file_path):
                    df_conditions[task][sess][run] = pd.read_csv(csv_file_path)
                else:
                    print(f"Warning: CSV file not found: {csv_file_path}")

            except (OSError, KeyError, ValueError) as e:
                print(f"Error processing {h5_file_path}: {e}")
```

```python
# Within each session, for each task, calculate the contrast beta between the encoding and delay period
delay_task_betas, delay_task_df = filter_task_betas(task_betas, df_conditions, phase="delay", first_delay_only=False)
encoding_task_betas, encoding_task_df = filter_task_betas(task_betas, df_conditions, phase="encoding", first_delay_only=False)
```

## Identify significant regions [delay VS. encoding]

```python
# Flag to toggle significance testing
perform_significance_test = True  # Set to False to bypass significance testing

# Concatenate betas across all sessions and tasks, averaging within each brain region
all_delay_betas = beta_concatenation(delay_task_betas)
all_encoding_betas = beta_concatenation(encoding_task_betas)

# Initialize data structures
num_voxels = glasser_atlas.shape[0]
t_stats_map = np.zeros((num_voxels, 1))
p_values_map = np.zeros((num_voxels, 1))
p_values_list = []

# Identify significant regions
significant_region_index = []
positive_significant_region_index = []
negative_significant_region_index = []

# Perform statistical analysis for each brain region
for region in range(1, 361):
    region_idx = np.where(glasser_atlas == region)[0]

    # Compute average beta values within the region
    region_avg_encoding_beta = np.mean(all_encoding_betas[:, region_idx], axis=-1)
    region_avg_delay_beta = np.mean(all_delay_betas[:, region_idx], axis=-1)

    if region_avg_encoding_beta.shape[-1] > 0:
        # Paired t-test between encoding and delay phases
        t_stat, p_value = stats.ttest_rel(region_avg_encoding_beta, region_avg_delay_beta)

        # Store test results
        t_stats_map[region_idx] = t_stat
        p_values_map[region_idx] = p_value
        p_values_list.append(p_value)

        # Track significant regions before FDR correction
        if p_value < 0.05:
            significant_region_index.append(region)
            if t_stat > 0:
                positive_significant_region_index.append(region)
            else:
                negative_significant_region_index.append(region)

```

```python
# FDR correction
if perform_significance_test:
    significant_region_index.clear()
    positive_significant_region_index.clear()
    negative_significant_region_index.clear()

    # Convert p-values to a NumPy array for processing
    p_values_array = np.array(p_values_list)
    print(f"Total number of p-values: {p_values_array.shape[0]}")

    # Apply FDR correction using Benjamini-Hochberg procedure
    _, fdr_corrected_p_values, _, _ = multipletests(p_values_array, alpha=0.05, method='fdr_bh')

    # Assign corrected p-values back to brain regions
    fdr_index = 0
    for region in range(1, 361):
        region_idx = np.where(glasser_atlas == region)[0]

        if fdr_corrected_p_values[fdr_index] < 0.05:  # Significant after FDR correction
            significant_region_index.append(region)
            if t_stats_map[region_idx].mean() > 0:
                positive_significant_region_index.append(region)
            else:
                negative_significant_region_index.append(region)
        else:
            t_stats_map[region_idx] = 0  # Set non-significant regions to 0

        fdr_index += 1
else:
    print("Significance testing is skipped. Using raw t-statistics.")
```

```python
# Plot the brain surface with corrected t-values
make_brain_surface_plot(
    np.squeeze(t_stats_map), 
    f"All tasks, all sessions t-value distribution (excluding session {excluded_session})"
)

# Output results
if perform_significance_test:
    print(f"Number of significant regions after FDR correction: {len(significant_region_index)}")
else:
    print(f"Number of significant regions (without correction): {len(significant_region_index)}")
    print("Significance testing was skipped. No significant regions identified.")
```

```python
# Convert lists to arrays and save
np.savez("results/glm_betas_encoding_delay_full_TR_betas_alltasks_contrast_significant_regions.npz",
         positive=positive_significant_region_index,
         significant=significant_region_index,
         negative=negative_significant_region_index)

print("Lists saved successfully.")


```

## Feature Decoding Analysis

```python
# decoding given feature across all tasks, leave one session out

# Flag to toggle significance testing
perform_significance_test = True  # Set to False to bypass significance testing

decoding_feature = "category"
phase = "delay"
classifier='distance'
feature_normalization = True

# todo: add subsampling by the filtered significant regions

# Initialize maps and lists
significant_decodable_region_index = []

regionwise_acc = np.zeros((64984, 1))
t_stats_map = np.zeros((64984, 1))
p_values_map = np.zeros((64984, 1))
p_values_list = []  # Collect all p-values for FDR correction. Q: FDR correction on all brain regions?


```

```python
import time

# Iterate through each brain region
for region in range(1, 361):
    print(f"Current region: {region}")
    # Time beta concatenation
    start_time = time.time()
    region_idx = np.where(glasser_atlas == region)[0]

    # Collect accuracy for each fold
    fold_accuracies = []

    for exclude_ses in sessions:
        exclude_ses = [exclude_ses]

        
        all_delay_betas = beta_concatenation(delay_task_betas, exclude_ses)
        all_encoding_betas = beta_concatenation(encoding_task_betas, exclude_ses)
        all_df = df_concatenation(delay_task_df, exclude_ses)
        beta_concat_time = time.time() - start_time


        # Time data selection
        start_time = time.time()
        if phase == "delay":
            data = all_delay_betas[:, region_idx]
        else:
            data = all_encoding_betas[:, region_idx]
        labels = all_df[decoding_feature].to_numpy()
        data_selection_time = time.time() - start_time

        # Ensure that the region has data
        if data.shape[-1] != 0:
            # Time Stratified K-Fold cross-validation
            start_time = time.time()
            skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
            for train_index, test_index in skf.split(data, labels):
                trainset, testset = data[train_index], data[test_index]
                trainlabels, testlabels = np.array(labels)[train_index], np.array(labels)[test_index]
                accuracy = decoding(trainset, testset, trainlabels, testlabels, classifier=classifier, confusion=False, feature_normalization=feature_normalization)
                fold_accuracies.append(np.mean(accuracy))
            skf_time = time.time() - start_time

            # Time t-test and map updates
            start_time = time.time()
            t_stat, p_val = ttest_1samp(fold_accuracies, popmean=0.5, alternative='greater')
            t_stats_map[region_idx] = t_stat
            p_values_map[region_idx] = p_val
            ttest_time = time.time() - start_time

    # Time region-wise accuracy calculation
    regionwise_acc[region_idx] = np.mean(fold_accuracies)
    accuracy_calc_time = time.time() - start_time
    print(f"Region-wise accuracy calculation time: {accuracy_calc_time:.4f} seconds")
    print(f"Mean accuracy for region {region}: {np.mean(fold_accuracies):.4f}")
```
